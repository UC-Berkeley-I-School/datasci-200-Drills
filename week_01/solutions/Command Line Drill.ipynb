#### First let's see what the current folder we are located in:

- The command to do this is `pwd`
- Type that into the terminal!

#### From this folder - do the following commands:

1) Make a new directory called **first_drill**<br>
2) Make a new file called **notes.txt**<br>
3) Using the editor VIM, add the first line to notes.txt - "Testing out the VIM editor!"<br>
   - You can start VIM by the command `vim notes.txt`
   - Make sure you are inserting into the file (Hit the INSERT key until it says -- INSERT -- in the bottom left corner)
   - Remember to close VIM by hitting ESC then typing: ':wq' and hitting enter<br>

4) Now from the command line, add this new line to **notes.txt** - "Adding text from the command line."<br>
5) From the command line, look at the contents of **notes.txt**<br>
6) Using relative paths, move **notes.txt** to the folder **first_drill**<br>
7) Without using `cd`, make a new folder in the **first_drill** folder called **extra_notes**<br>
8) Using relative paths, copy **notes.txt** to the new folder **extra_notes** made in 7 above<br>
9) Rename the **notes.txt** in the **first_drill/extra_notes** folder to **copy_of_notes.txt**<br>

## Below is the list of commands:

1) `mkdir first_drill`<br>
2) `touch notes.txt`<br>
3) `vim notes.txt` 
   - Typed in: Testing out the VIM editor!
   - hit ESC, ':wq' to write to the file and quit back to the command line<br>
   
4) `echo "Adding text from the command line." >> notes.txt`<br>
5) `cat notes.txt`<br>
6) `mv notes.txt ./first_drill`<br>
7) `mkdir ./first_drill/extra_notes`<br>
8) `cp ./first_drill/notes.txt ./first_drill/extra_notes/`<br>
9) `mv ./first_drill/extra_notes/notes.txt ./first_drill/extra_notes/copy_of_notes.txt`

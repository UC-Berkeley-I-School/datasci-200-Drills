### A note about Theta Notation ($\Theta$)

Theta notation is what's known as a tight asymptotic bound, by tight we mean it defines both an upper and a lower bound for the running time as the input size gets arbitrairly large.<br>
Essentially stating that an algorithm is $\Theta(f(n))$, where $f(n)$ is any function of the input size, bounds the running time of that algorithm between two multiples of $f(n)$, or more formally:<br>
Algorithm is $\Theta(f(n)) \ \ \rightarrow \ \ c_1 \cdot f(n) \leq \text{running time} \leq c_2 \cdot f(n) \ \ \ \text{for large } n$<br>

In this drill we will be concerned with defining $f(n)$ and we won't worry about deriving the constants $c_1$ and $c_2$.  

### Constant Time $\Theta$(1)

For $\Theta$(1), there are constants $c_1$ and $c_2$, such that the runtime is bounded by $c_1$ and $c_2$ regardless of the size of the input.<br>

Many operations in python are $\Theta$(1). For example, adding two integers, popping the last element of a list, or accessing a dictionary.

```python
list_1 = [i for i in range(1000)] 
list_2 = [i for i in range(100)]

dict_1 = {chr(i):i for i in range(33,127)}
dict_2 = {chr(i):i for i in range(33,64)}

# Each of the operations below run in constant time, regardless of the size of the input. (ignoring the time needed to declare and initialize the variables).  

#Adding integers is Theta(1)
x = 1000 + 2000
print(x)
print("-"*20)

#Accessing an elemnt of a list is Theta(1), regardless of the size of the array
x = list_1[952]
y = list_2[81]
print(x)
print(y)
print("-"*20)

#The operation below is also Theta(1)
z = list_1[753] + list_2[25]
print(z)
print("-"*20)

#Popping the last time of a list is Theta(1)
x = list_1.pop()
y = list_2.pop()
print(x)
print(y)
print("-"*20)

#Getting an item from a dictionary is Theta(1), regardless of the size of the dictionary
x = dict_1["A"]
y = dict_1["!"]
print(x)
print(y)
```

## Linear Time $\Theta$(n)

$\Theta$(n) describes an algorithm whose performance will grow linearly and in direct proportion to the size of the input data set. For $\Theta$(n), there are constants $c_1$ and $c_2$, such that the runtime is bounded by $c_1 \cdot n$ and $c_2 \cdot n$, where n measures the size of the input. For example, finding the number of occurrences of a particular integer in an unsorted array of n integers by the following algorithm is $\Theta$(n).

Another example below also demonstrates how Theta notation has to account for the worst-case performance scenario; a matching string could be found during any iteration of the for loop and the function would return early, but Theta notation will need to also consider the upper limit where the algorithm will perform the maximum number of iterations.

```python
# Here the upper limit is the length of the list.

def containsValue(aList, findMe):
    for x in aList:
        print("x = ",x," compared to ",findMe)
        if x == findMe:
            return True
    return False

print(containsValue([1,3,6,2,77,5,4],5))
```

The complexity of conditionals depends on what the condition is. In this example, we have an if statement. The analysis of the 
runtime of a conditional is highly dependent upon what the conditionalâ€™s condition actually is; checking if one character 
is equal to another is a constant-time operation, so this example is linear with respect to the size of a str. 

So, if we let n = |a str|, this function is $\Theta$(n).

```python
def count_ts(a_str):
    count = 0
    for char in a_str:
        if char == 't':
            count += 1
    return count

print(count_ts("crazy cat"))
```

### Cubic time \[$\Theta$(n^3)\]:  

The below example is $\Theta$(n^3) or cubic. Often cubic complexity is encountered when we have to manipulate matrices. This can grow in size awfully fast ... 

```python
# Theta(N3)is merely Theta(N2) with another exponent. 
# this snippet shows the difference by changing the exponent.

def TN3(input_):
    count = 0
    for i in input_:
        count += 1
        for j in input_:
            count += 1
            for k in input_:
                count +=1
    return 1 + count

print(TN3([1,2,3,4,5]))
```

### Logarithmic time \[$\Theta$(log(n))\]:  

$\Theta$(logN), or logarithmic complexity, exhibits an increase in execution time proportional the log of the input size; this means you can exponentially increase your input size without an exponential increase in execution time. This comes from the properties of logarithms, where log(n^x) = xlog(n). Logarithmic time is often encountered in search algorithms.

```python
def TlogN(input_):
    count = 0
    high = len(input_) - 1
    while high//2 != 0:
        high = high//2
        count +=1
    return count

print(TlogN([1,2,3,4,5]))
```

### Log-linear time \[$\Theta$(nlog(n))\]:  

$\Theta$(NlogN), or log-linear, grows faster than $\Theta$(n) but is still slower than $\Theta$(n^2). You might encounter log-linear complexities in sorting algorithms later this week.

```python
def TNlogN(input_, count=0):
    length = len(input_)
    count += 1
    if length <= 1:
        return count
    else:
        return TNlogN(input_[:length//2], count) + TNlogN(input_[length//2:], count)
    
print(TNlogN([1,2,3,4,5]))
```

### Testing against commong growth functions
Below we define a function, test_theta(),which tests a function passed as an argument against the common growth functions which we just went over.

```python
#pandas is a package which we will learn about later in the course
#we use it here because it makes it easier to print clean tables.

import pandas as pd 
import math

#format of float number display in pandas
pd.options.display.float_format = "{:,.2f}".format

def test_theta(func, no_of_doubles):
    """Runs trials doubling size of input list until no_of_doubles reached.
    Returns dataframe with ratios of function number of iterations to common growth functions.
    
    Note: func is required to return number of iterations for each run of func
    """
    
    out = []
    #compute table
    for i in range(1,no_of_doubles+1):
        n = 2**i #input size
        func_n = func([j for j in range(n)])
        out.append([n,
                    func_n,
                    func_n / math.log(n, 2),
                    func_n / n,
                    func_n / (n * math.log(n, 2)),
                    func_n / n**2,
                    func_n / n**3,
                   ]
                  )    
    df = pd.DataFrame(out, columns = ["n", "func", "func/log2(n)", "func/n", "func/(nlog2(n))", "func/n^2", "func/n^3"])
    
    return df
```

Notice how some ratios converge to zero, some converge to a constant, and others explode for different complexity functions.

```python
test_theta(TlogN,20)
```

For this drill, let's continue adding to our chart from the last drill - looking at the average rainfall in Singapore over the last 5 years (2016-2020). The dataset is in this folder named: `Singapore_Monthly_Rainfall.csv`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 1000)
pd.options.display.float_format = '{:,.2f}'.format
%matplotlib inline
```

```python
# Loading up the cleaned dataset from the last drill

sing_final = pd.read_csv('singapore_rainfall_cleaned.csv')
```

Below is the chart that was made in the previous drill - as you can see it is missing some important items such as the titles and labels. Also I would prefer the x-axis (months) to show all of the months as well as the name of the month rather than the number.

```python
plt.figure(figsize=(20,10))

plt.plot(sing_final['Month'], sing_final['Rainfall_2016'], label = "2016", linestyle="--")
plt.plot(sing_final['Month'], sing_final['Rainfall_2017'], label = "2017")
plt.plot(sing_final['Month'], sing_final['Rainfall_2018'], label = "2018")
plt.plot(sing_final['Month'], sing_final['Rainfall_2019'], label = "2019", color = 'purple')
plt.plot(sing_final['Month'], sing_final['Rainfall_2020'], label = "2020", color = 'red')

plt.legend(loc = 'upper left')
plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.plot(sing_final['Month'], sing_final['Rainfall_2016'], label = "2016", linestyle="--")
plt.plot(sing_final['Month'], sing_final['Rainfall_2017'], label = "2017")
plt.plot(sing_final['Month'], sing_final['Rainfall_2018'], label = "2018")
plt.plot(sing_final['Month'], sing_final['Rainfall_2019'], label = "2019", color = 'purple')
plt.plot(sing_final['Month'], sing_final['Rainfall_2020'], label = "2020", color = 'red')

# Adding in the axis labels and title:
plt.xlabel('Month')
plt.ylabel('Total Rainfall')
plt.title('Comparison of 2016-2020 Monthly Rainfall in Singapore')

plt.legend(loc = 'upper left')
plt.show()
```

```python
# fixing the months so the x-axis displays the name

plt.figure(figsize=(20,10))

plt.plot(sing_final['Month'], sing_final['Rainfall_2016'], label = "2016", linestyle="--")
plt.plot(sing_final['Month'], sing_final['Rainfall_2017'], label = "2017")
plt.plot(sing_final['Month'], sing_final['Rainfall_2018'], label = "2018")
plt.plot(sing_final['Month'], sing_final['Rainfall_2019'], label = "2019", color = 'purple')
plt.plot(sing_final['Month'], sing_final['Rainfall_2020'], label = "2020", color = 'red')

# Adding the monthly labels to the x-axis ticks
labels = ['Jan','Feb','Mar','Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(sing_final['Month'], labels, rotation=45)

# Adding in the axis labels and title:
plt.xlabel('Month', fontsize=18)
plt.ylabel('Total Rainfall', fontsize=18)
plt.title('Comparison of 2016-2020 Monthly Rainfall in Singapore', fontsize=24)

plt.legend(loc = 'upper left')
plt.show()
```

### Insights:
    
Now that the chart is done let's look for insights that can be gained from it.
- The rainiest month in most years (all but 2018) looks to be December
- The trend in the July/Aug/Sep month seemed to be less and less year over year (2016-2019) but came up in 2020
- 3 months of low rain total in 2019 might be something to look into (why did it happen? what were the weather patterns?)
